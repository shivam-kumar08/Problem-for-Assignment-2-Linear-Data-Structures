```python
#1. Delete the elements in a linked list whose sum is equal to zero
class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def delete_zero_sum_sublists(self):
        dummy = ListNode(0)
        dummy.next = self.head
        prefix_sum = {0: dummy}
        current = self.head
        current_sum = 0

        while current:
            current_sum += current.value
            if current_sum in prefix_sum:
                prev = prefix_sum[current_sum]
                temp = prev.next
                while temp != current:
                    current_sum += temp.value
                    del prefix_sum[current_sum]
                    temp = temp.next
                prev.next = current.next
            else:
                prefix_sum[current_sum] = current
            current = current.next

    def print_list(self):
        current = self.head
        while current:
            print(current.value, end=" -> ")
            current = current.next
        print("None")

# Example usage:
linked_list = LinkedList()
linked_list.head = ListNode(3)
linked_list.head.next = ListNode(4)
linked_list.head.next.next = ListNode(-7)
linked_list.head.next.next.next = ListNode(5)
linked_list.head.next.next.next.next = ListNode(-6)
linked_list.head.next.next.next.next.next = ListNode(6)
linked_list.head.next.next.next.next.next.next = ListNode(2)
linked_list.head.next.next.next.next.next.next.next = ListNode(2)

print("Original Linked List:")
linked_list.print_list()

linked_list.delete_zero_sum_sublists()
print("Linked List after deleting zero-sum sublists:")
linked_list.print_list()
```

```python
#2 Reverse a linked list in groups of given siz
class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def reverse_in_groups(self, head, k):
        current = head
        next = None
        prev = None
        count = 0

        # Count the number of nodes in the current group
        temp = head
        while temp and count < k:
            temp = temp.next
            count += 1

        # If there are at least k nodes in the current group, reverse them
        if count == k:
            while count > 0 and current is not None:
                next = current.next
                current.next = prev
                prev = current
                current = next
                count -= 1

            if next is not None:
                head.next = self.reverse_in_groups(next, k)

            return prev
        else:
            return head

    def print_list(self):
        current = self.head
        while current:
            print(current.value, end=" -> ")
            current = current.next
        print("None")

# Example usage:
linked_list = LinkedList()
linked_list.head = ListNode(1)
linked_list.head.next = ListNode(2)
linked_list.head.next.next = ListNode(3)
linked_list.head.next.next.next = ListNode(4)
linked_list.head.next.next.next.next = ListNode(5)
linked_list.head.next.next.next.next.next = ListNode(6)
linked_list.head.next.next.next.next.next.next = ListNode(7)
linked_list.head.next.next.next.next.next.next.next = ListNode(8)

print("Original Linked List:")
linked_list.print_list()

k = 3
linked_list.head = linked_list.reverse_in_groups(linked_list.head, k)
print(f"Linked List after reversing in groups of {k}:")
linked_list.print_list()

```

```python
#3 Merge a linked list into another linked list at alternate positions.
class ListNode:
    def __init__(self, value):
        self.value = value
        self.next = None

class LinkedList:
    def __init__(self):
        self.head = None

    def merge_alternate(self, l2):
        current1 = self.head
        current2 = l2.head

        while current1 is not None and current2 is not None:
            next1 = current1.next
            next2 = current2.next

            current1.next = current2
            current2.next = next1

            current1 = next1
            current2 = next2

    def print_list(self):
        current = self.head
        while current:
            print(current.value, end=" -> ")
            current = current.next
        print("None")

# Example usage:
linked_list1 = LinkedList()
linked_list1.head = ListNode(1)
linked_list1.head.next = ListNode(3)
linked_list1.head.next.next = ListNode(5)

linked_list2 = LinkedList()
linked_list2.head = ListNode(2)
linked_list2.head.next = ListNode(4)
linked_list2.head.next.next = ListNode(6)

print("Original Linked List 1:")
linked_list1.print_list()

print("Original Linked List 2:")
linked_list2.print_list()

linked_list1.merge_alternate(linked_list2)
print("Merged Linked List:")
linked_list1.print_list()
```

```python
#4In an array, Count Pairs with given sum
class PairCounter:
    def __init__(self, arr):
        self.arr = arr

    def count_pairs_with_sum(self, target_sum):
        count = 0
        num_counts = {}

        for num in self.arr:
            complement = target_sum - num
            if complement in num_counts:
                count += num_counts[complement]
            num_counts[num] = num_counts.get(num, 0) + 1

        return count

# Example usage:
arr = [1, 2, 3, 4, 5, 6, 7, 8, 9]
target_sum = 10
pair_counter = PairCounter(arr)
count = pair_counter.count_pairs_with_sum(target_sum)
print(f"Number of pairs with sum {target_sum}: {count}")
```

```python
#5Find duplicates in an array
class DuplicatesFinder:
    def __init__(self, arr):
        self.arr = arr

    def find_duplicates(self):
        duplicates = set()
        seen = set()

        for num in self.arr:
            if num in seen:
                duplicates.add(num)
            seen.add(num)

        return list(duplicates)

# Example usage:
arr = [1, 2, 3, 4, 2, 5, 6, 1, 7, 8, 8, 9]
duplicates_finder = DuplicatesFinder(arr)
duplicates = duplicates_finder.find_duplicates()
print("Duplicates in the array:", duplicates)
```

```python
#6Find the Kth largest and Kth smallest number in an array
class KthLargestSmallestFinder:
    def __init__(self, arr):
        self.arr = arr

    def find_kth_largest(self, k):
        sorted_arr = sorted(self.arr, reverse=True)
        if 1 <= k <= len(sorted_arr):
            return sorted_arr[k - 1]

    def find_kth_smallest(self, k):
        sorted_arr = sorted(self.arr)
        if 1 <= k <= len(sorted_arr):
            return sorted_arr[k - 1]

# Example usage:
arr = [12, 11, 13, 5, 6, 7]
k = 3
finder = KthLargestSmallestFinder(arr)
kth_largest = finder.find_kth_largest(k)
kth_smallest = finder.find_kth_smallest(k)
print(f"{k}th Largest element: {kth_largest}")
print(f"{k}th Smallest element: {kth_smallest}")
```

```python
#7Move all the negative elements to one side of the array
class NegativeMover:
    def __init__(self, arr):
        self.arr = arr

    def move_negatives(self):
        left = 0
        right = len(self.arr) - 1

        while left <= right:
            if self.arr[left] < 0 and self.arr[right] < 0:
                left += 1
            elif self.arr[left] > 0 and self.arr[right] < 0:
                self.arr[left], self.arr[right] = self.arr[right], self.arr[left]
                left += 1
                right -= 1
            elif self.arr[left] > 0 and self.arr[right] > 0:
                right -= 1
            else:
                left += 1
                right -= 1

# Example usage:
arr = [-12, 11, -13, -5, 6, -7, 5, -3]
negative_mover = NegativeMover(arr)
negative_mover.move_negatives()
print("Array after moving negatives to one side:", arr)
```

```python
#8 Reverse a string using a stack data structure
class StringReverser:
    def __init__(self, string):
        self.string = string

    def reverse(self):
        stack = []
        for char in self.string:
            stack.append(char)
        reversed_string = ""
        while stack:
            reversed_string += stack.pop()
        return reversed_string

# Example usage:
input_string = "Hello, World!"
reverser = StringReverser(input_string)
reversed_string = reverser.reverse()
print("Reversed string:", reversed_string)
```

```python
#9 Evaluate a postfix expression using stack
class PostfixEvaluator:
    def __init__(self, expression):
        self.expression = expression

    def evaluate(self):
        stack = []
        operators = set(['+', '-', '*', '/'])

        for char in self.expression:
            if char.isdigit():
                stack.append(int(char))
            elif char in operators:
                operand2 = stack.pop()
                operand1 = stack.pop()
                if char == '+':
                    stack.append(operand1 + operand2)
                elif char == '-':
                    stack.append(operand1 - operand2)
                elif char == '*':
                    stack.append(operand1 * operand2)
                elif char == '/':
                    stack.append(operand1 / operand2)

        if len(stack) == 1:
            return stack[0]
        else:
            raise ValueError("Invalid postfix expression")

# Example usage:
postfix_expression = "23*5+"
evaluator = PostfixEvaluator(postfix_expression)
result = evaluator.evaluate()
print("Result of postfix expression:", result)
```

```python
#10 Implement a queue using the stack data structure
class StackQueue:
    def __init__(self):
        self.stack1 = []
        self.stack2 = []

    def enqueue(self, value):
        self.stack1.append(value)

    def dequeue(self):
        if not self.stack2:
            if not self.stack1:
                raise IndexError("Queue is empty")
            while self.stack1:
                self.stack2.append(self.stack1.pop())
        return self.stack2.pop()

# Example usage:
queue = StackQueue()
queue.enqueue(1)
queue.enqueue(2)
queue.enqueue(3)

print(queue.dequeue())  # Should print 1
print(queue.dequeue())  # Should print 2

queue.enqueue(4)
print(queue.dequeue())  # Should print 3
print(queue.dequeue())  # Should print 4
```
